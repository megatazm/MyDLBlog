# Perceptron
*Perceptron* adalah salah satu *architecture ANN* termudah, dicipta pada tahun 1957 oleh Frank Rosenblatt. Ia berdasarkan *neuron* (lihat Gambar di bawah) yang disebut Treshold Logic Unit (TLU), ataupun *Logic Treshold Unit* (LTU). Nilai input dan output adalah nombor (bukan nilai on / off binari), dan setiap sambungan di antara input nod dan output nod diberikan pemberat atau *weight* Secara analoginya, lagi besar weight maka lagi tebal lah wayar sambungannya.
